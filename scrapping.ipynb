```python
import pandas as pd
import numpy as np 
```

```python
url = "http://www.econstats.com/blscu/cu_itemSAMm.htm"
```

```python
tables = pd.read_html(url)
df = tables[13]

for i in range(6):
    del df[i]


df.head(5)
```

```python
list = ["CUURA318","CUURA319", "CUURA320", "CUURA421", "CUURA422", "CUURA423", "CUURA101",
        "CUURA102", "CUURA103"]

result = []
for code in list:
    name = code + "SAM"
    data = df[df[6] == name]
    result.append(data)


```

```python
final = pd.concat(result)
final = final.reset_index(drop = True)

final.head(2)
```

```python
for i in range(7,13):
    del final[i]

final
```
