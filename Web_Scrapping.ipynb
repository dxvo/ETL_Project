# Web Scrapping to obtain Medical Consumer Prince Index

```python
import pandas as pd
import numpy as np
import config
```

```python
#link to source page 
med_url = "http://www.econstats.com/blscu/cu_itemSAMm.htm"
```

```python
#Respond from web using pd.read_html 
tables = pd.read_html(med_url)
web_data = tables[13] 
for i in range(6):
    del web_data[i]
web_data.head(3)
```

```python
#Code for list of interest area 
area_list = ["CUURA318","CUURA319", "CUURA320", "CUURA421", "CUURA422", "CUURA423", "CUURA101",
        "CUURA102", "CUURA103"]

result = []
for code in area_list:
    name = code + "SAM"
    data = web_data[web_data[6] == name]
    result.append(data)
```

```python
final_df = pd.concat(result)
#final = final.reset_index(drop = True)
final_df.head(2)
```

```python
#Cleaning - Deleting unnecessary cols 
for i in range(6,13):
    del final_df[i]
final_df
```

```python
final_df = final_df.T
final_df.head(5)
```

```python
final_df.columns = area_list
final_df.head(3)
```

```python
#Creat of range of date that scrapped
dates = pd.date_range('2009-04-01','2012-05-01', freq='1M')- pd.offsets.MonthBegin(1)
#dates = dates.strftime("%Y-%m-%d").tolist()
dates
```

```python
#Set index of final df to dates 
final_df.index = dates
final_df.head(5)
```

# Load Data to MYSQL

```python
# Imports the method used for connecting to DBs
from sqlalchemy import create_engine
# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()
```

```python
# Create Engine and Pass in MySQL Connection                
MySQL_root_PW = config.password
MySQL_db = 'ETL'
engine = create_engine("mysql://root:"+MySQL_root_PW+"@localhost/"+MySQL_db)
```

```python
#load data 
final_df.to_sql("Medical", 
                  con = engine,
                  index = True, 
                  if_exists='replace')
```

```python
#test by reading
pd.read_sql("Medical", engine)
```
